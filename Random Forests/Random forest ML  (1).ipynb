```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

import numpy as np

# My fake data: [age, income], target (1 = buy, 0 = no buy)
X = np.array([[25, 30000], [30, 50000], [20, 20000], [35, 60000], [28, 45000]])
y = np.array([0, 1, 0, 1, 1])

# Build 3 trees with Bagging
trees = []
for i in range(3):
    # Bootstrap sample
    indices = np.random.choice(len(X), size=len(X), replace=True)
    X_sample = X[indices]
    y_sample = y[indices]

    # Train a tree
    tree = DecisionTreeClassifier()
    tree.fit(X_sample, y_sample)
    trees.append(tree)

# Predict for a new person: age 27, income 40000
new_person = np.array([[27, 40000]])
votes = [tree.predict(new_person)[0] for tree in trees]
print("Tree votes:", votes)
final_guess = max(set(votes), key=votes.count)  # Majority vote
print("Bagging says:", "Buy" if final_guess == 1 else "No buy")

```

```python
# Same data
X = np.array([[25, 30000], [30, 50000], [20, 20000], [35, 60000], [28, 45000]])
y = np.array([0, 1, 0, 1, 1])

# Train a Random Forest with 3 trees
rf = RandomForestClassifier(n_estimators=3, random_state=42)
rf.fit(X, y)

# Predict for our new person
new_person = np.array([[27, 40000]])
guess = rf.predict(new_person)
print("Random Forest says:", "Buy" if guess[0] == 1 else "No buy")
```

```python
from sklearn.metrics import accuracy_score

# Real vs. predicted
real = [1, 0, 0, 1]
predicted = [1, 0, 1, 1]
accuracy = accuracy_score(real, predicted)
print("Accuracy:", accuracy * 100, "%")
```
