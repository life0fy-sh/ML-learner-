```python
# Import necessary libraries
import numpy as np #numpy is a library for numerical computing
from sklearn.linear_model import LinearRegression # scikit-learn is a library for machine learning
import matplotlib.pyplot as plt # matplotlib is a library for plotting
```

```python
# read csv file
import pandas as pd
data = pd.read_csv('real_estate_price_size.csv')

print(data.head())
```

```python
# Assign the data to X and y
X = data['size'] # X is the feature we are using to predict y
Y = data['price'] # y is the target variable we are trying to predict
```

```python
# Reshape the data
X = X.values.reshape(-1,1)
Y = Y.values.reshape(-1,1)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X, Y.reshape(-1, 1))  # Reshape Y to be 2-dimensional
```

```python
# Get the slope (m) and intercept (b)
m = model.coef_[0]  # Slope (how steep the line is)
b = model.intercept_  # Intercept (where the line crosses the y-axis)
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
```

```python
# CREATE a scatter plot
plt.scatter(X, Y, color='blue')
# SET the title and labels
plt.title('Real Estate Prices')
plt.xlabel('Size')
plt.ylabel('Price')
# PLOT the regression line
plt.plot(X, m*X + b, color='red')
plt.show()
```
