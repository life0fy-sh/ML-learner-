```python
# Import necessary libraries
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Step 1: Prepare the data
# Features: Weight and Price (X)
X = np.array([
    [201, 150],  # Samsung Galaxy A14
    [206, 999],  # iPhone 14 Pro
    [188, 200],  # Xiaomi Redmi Note 12
    [197, 599],  # Google Pixel 7
    [150, 400],  # Realme
    [220, 440],  # Samsung Galaxy A16
    [250, 1460], # iPhone 15 Pro
    [230 ,1099], # Google Pixel 7
    [260,1200],  # iPhone 16 Pro
    [260,1190]   # Google pixel 8
    ])

# Labels: Categories (y) - 0 for Cheap, 1 for Expensive
y = np.array([0, 1, 0, 1, 0, 0,1 , 1, 1, 1])

# Step 2: Create and train the KNN model
knn = KNeighborsClassifier(n_neighbors=3)  # K = 3
knn.fit(X, y)

# Step 3: Predict the category of OnePlus Nord CE 3 (184, 300)
oneplus_nord = np.array([[210, 200]])
prediction = knn.predict(oneplus_nord)

Oppp_reno = np.array([[210,999]])
prediction = knn.predict(Oppp_reno)

# Step 4: Interpret the result
categories = {0: "Cheap", 1: "Expensive"}
print(f"Predicted category for OnePlus Nord CE 3: {categories[prediction[0]]}")
print(f"Predicted Category for Oppo reno : {categories[prediction[0]]}")
```
